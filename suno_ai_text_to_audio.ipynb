```python
import torch
import torchaudio
import numpy as np
import scipy
import os
from bark import generate_audio, preload_models

# Override torch.load to allow full model loading
original_load = torch.load

def custom_load(*args, **kwargs):
    kwargs["weights_only"] = False  # Allow loading full model
    return original_load(*args, **kwargs)

torch.load = custom_load  # Temporarily override torch.load

# Now, load Bark models
preload_models()

# Restore the original torch.load after loading models
torch.load = original_load

# Define the voice presets for different emotions
emotion_voice_presets = {
    "angry": "v2/en_speaker_6",       # Intense voice
    "excited": "v2/en_speaker_3",      # High-energy voice
    "shouting": "v2/en_speaker_5",     # Loud and energetic voice
    "surprised": "v2/en_speaker_8",    # Sudden high pitch
    "aggressive": "v2/en_speaker_9",   # Harsh tone
    "fearful": "v2/en_speaker_7",      # Nervous, trembling voice
    "whispering": "v2/en_speaker_2"    # Soft whisper
}

# List of text and emotions
comic_lines = [
    ("THE FINAL REPORT CARD! LET'S SEE", "excited"),
    ("AN E?! AN E IN ARITHMETIC?! Did I FAIL ARITHMETIC?!", "excited"),
    ("HAHA CHARLES GOT AN E! LOOK! CHARLES GOT AN E!", "excited"),
    ("HEY! COME SEE! HE GOT AN E! HE FAILED!", "shouting"),
    ("WHY DO I ONLY HAPPEN TO GET NOTICED WHEN I SLIP UP?", "whispering"),
    ("WINTER VACATION IS UPON US", "whispering"),
    ("I LOVE WINTER BREAK! I'LL SPEND ALL TWO WEEKS RELAXING AND READING COMIC BOOKS AND MAGAZINES ABOUT FOOTBALL AND HOCKEY", "excited"),
    ('"SILAS MARNER"?! WE HAVE TO READ "SILAS MARNER" OVER WINTER BREAK?!', "excited"),
    ("AAAH", "fearful")
]

# Directory to save output files
os.makedirs("generated_speech", exist_ok=True)

# Generate and save audio for each line
for i, (text, emotion) in enumerate(comic_lines):
    voice_preset = emotion_voice_presets.get(emotion, "v2/en_speaker_3")  # Default to excited
    audio_array = generate_audio(text, history_prompt=voice_preset)

    # Save audio file
    output_path = f"generated_speech/line_{i+1}_{emotion}.wav"
    scipy.io.wavfile.write(output_path, rate=24000, data=(audio_array * 32767).astype(np.int16))

    print(f"Saved: {output_path}")
```

```python
!pip install pydub
from pydub import AudioSegment
import os

# Directory where the generated .wav files are stored
output_dir = "generated_speech"
output_files = sorted([f for f in os.listdir(output_dir) if f.endswith(".wav")])

# Ensure there are files to merge
if not output_files:
    print("No audio files found to merge.")
else:
    # Load the first audio file as the base
    combined_audio = AudioSegment.from_wav(os.path.join(output_dir, output_files[0]))

    # Append the rest of the files in sequence
    for file in output_files[1:]:
        audio_segment = AudioSegment.from_wav(os.path.join(output_dir, file))
        combined_audio += audio_segment  # Concatenate

    # Export the final merged audio
    final_output_path = "combined_audio.wav"
    combined_audio.export(final_output_path, format="wav")

    print(f"Successfully merged {len(output_files)} files into: {final_output_path}")
```
