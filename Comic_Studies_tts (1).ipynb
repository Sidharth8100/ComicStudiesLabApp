```python
from TTS.api import TTS
import warnings

# Suppress warnings
warnings.filterwarnings("ignore", category=FutureWarning)

# Load XTTS model
tts = TTS(model_name="tts_models/multilingual/multi-dataset/xtts_v2", progress_bar=False).to("cpu")

# Path to reference speaker audio (change this to a real `.wav` file path)
speaker_wav = "reference.wav"  # Provide a valid WAV file of a speaker

# Comic sounds with emotions
comic_sounds = {
    "KABOOM!": "angry",
    "POW!": "excited",
    "BOOM!": "shouting",
    "ZAP!": "surprised",
    "BAM!": "aggressive",
    "CRASH!": "fearful",
    "WHOOSH!": "whispering"
}

# Ensure a speaker reference is provided
if not speaker_wav:
    raise ValueError("You must provide a valid 'speaker_wav' file for voice cloning.")

# Generate and save each sound
for text, emotion in comic_sounds.items():
    tts.tts_to_file(
        text=text,
        file_path=f"{emotion}.wav",
        language="en",
        emotion=emotion,
        speaker_wav=speaker_wav  # Provide speaker reference
    )
    print(f"Generated '{text}' with emotion: {emotion}")

print("All comic sounds generated successfully!")
```

```python
import os
from TTS.api import TTS
from pydub import AudioSegment

# Load the TTS model
tts = TTS(model_name="tts_models/multilingual/multi-dataset/xtts_v2", progress_bar=False).to("cpu")

# Reference speaker audio file
reference_audio = "reference.wav"  # Replace with the actual path

# Comic-style text split into lines with emotions
comic_lines = [
    ("""THE FINAL
  REPORT CARDI
  LET'S SEE""", "excited"),
    ("""AN E?!
  AN E IN ARITHMETIC?I
  Did I FAIL ARITHMETiC?!""" "shocked"),
    ("""HAHA CHARLES
  GOT AN E! LOOK!
  CHARLES GOT AN E!""", "shouting"),
    ("""HEY! COME SEE!
  HE GoT AN E! HE FAILED!
  """, "shouting"),
    ("""WhY DO I
  ONLY HAPPEN TO
  GET NOTICED WHEN
  I SLIP UP?""", "whispering"),
    ("""WINTER VACATION
  IS UPON US""", "whispering"),
    ("""ILOVE WINTER BREAK!
  TLL SPEND ALL Two
  WEEKS RELAXING AND
  READING COMIC BOOKS
  AND MAGAZINES ABOUT
  FOOTBALL AND HOCKEY""", "excited"),
    (""" SILAS MARNER"?!
  WE HAVE TO READ
  "SILAS MARNER
  OVER
  WINTER BREAK?!""", "suprised"),
    ("""AAAAAUGH""","shouting" )
]

# Directory for storing temporary audio files
output_dir = "comic_audio"
os.makedirs(output_dir, exist_ok=True)

# Generate individual audio files
audio_files = []
for i, (line, emotion) in enumerate(comic_lines):
    output_file = f"{output_dir}/line_{i}.wav"
    tts.tts_to_file(
        text=line,
        file_path=output_file,
        language="en",
        emotion=emotion,
        speaker_wav=reference_audio
    )
    audio_files.append(output_file)

# Merge all audio files into one
final_audio = AudioSegment.empty()
for file in audio_files:
    final_audio += AudioSegment.from_wav(file)

# Export the final concatenated file
final_output = "comic_speech.wav"
final_audio.export(final_output, format="wav")

print(f"Comic-style narration successfully saved as '{final_output}'.")
```
